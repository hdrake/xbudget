```python
import xarray as xr
import xbudget
import numpy as np
```

```python
ds = xr.Dataset({
    "advective_tendency": xr.DataArray(np.random.rand(3, 3), dims=("x", "y")),
}, coords={"x":[0,1,2], "y":[0,1,2]})

xbudget_dict = {
    "heat": {
        "rhs": {
            "sum": {
                "advection": {"var": "advective_tendency"},
                "var": None,
            },
            "var": None,
        }
    }
}
```

```python
xbudget.collect_budgets(ds, xbudget_dict)
```
